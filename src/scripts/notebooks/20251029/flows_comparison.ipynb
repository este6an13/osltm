
### Results and Observations

The following observations are **based on preliminary visual analysis** of these **small subset of stations**. These findings are exploratory and serve mainly to guide future, more systematic analysis.

1. **Station-Specific Patterns**

   * Different stations exhibit **distinct temporal patterns**, suggesting that they could potentially be **classified or clustered** according to their characteristic behaviors.

2. **Temporal Similarities**

   * For **U Nacional**, patterns appear **similar on Tuesdays and Fridays**, indicating recurring weekday behavior.
   * The same station also shows **consistent patterns across months** and **across weeks within a month**, suggesting temporal stability at multiple scales.

3. **Holiday Effects**

   * **Calle 100** shows **notable differences between holiday and non-holiday days**, reflecting the impact of calendar events on passenger behavior.
   * **U Nacional** shows differences between vacation and non-vacation days, mostly in the magnitude of the peaks.

4. **Year-to-Year Consistency**

   * For **Calle 100**, patterns appear **similar between 2024 and 2025**, suggesting that certain structural or behavioral trends persist over time.
